```python
#streamlit already in requirements.txt; no pip
import streamlit as st
import pandas as pd
import numpy as np
import requests
import json
```

```python
#using the BLS public API Signature to retreive five datasets
api = "https://api.bls.gov/publicAPI/v1/timeseries/data/"

payload = {"seriesid":["CES0500000003","EIUIQ", "EIUIR","CES0000000001","LNS14000000"]}

response = requests.post(api, json=payload)
```

```python
#filtering through nested dictionary - layers 1 and 2
data = response.json()
data = data['Results']
data = data['series']
```

```python
#filtering through nested dictionary - layer 3
nonfarm = data[3]
avgHourly = data[0]
exportIndex = data[1]
importIndex = data[2]
unempRate = data[4]
```

```python
#filtering through nested dictionary - layer 4
nonfarm = nonfarm["data"]
avgHourly = avgHourly["data"]
exportIndex = exportIndex["data"]
importIndex = importIndex["data"]
unempRate = unempRate["data"]
```

```python
#drop columns
nonfarm = pd.DataFrame(nonfarm).drop(['footnotes', 'latest'], axis = 1)
avgHourly = pd.DataFrame(avgHourly).drop(['footnotes', 'latest'], axis = 1)
exportIndex = pd.DataFrame(exportIndex).drop(['footnotes', 'latest'], axis = 1)
importIndex = pd.DataFrame(importIndex).drop(['footnotes', 'latest'], axis = 1)
unempRate = pd.DataFrame(unempRate).drop(['footnotes', 'latest'], axis = 1)
```

```python
#save as CSV files
nonfarm.to_csv("nonfarm.csv")
avgHourly.to_csv("avgHourly.csv")
exportIndex.to_csv("exportIndex.csv")
importIndex.to_csv("importIndex.csv")
unempRate.to_csv("unempRate.csv")
```
